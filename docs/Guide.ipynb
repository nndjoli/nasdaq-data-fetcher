```python
import NSDQ
import BIS_CBPR
```

```python
r = BIS_CBPR.CentralBankPolicyRates('United States') #Get the Central Bank Policy Rate for the United States.
```

```python
AssetObject = NSDQ.Asset('AAPL','stocks') #Create an instance of the Asset class for Apple Inc.
```

```python
AssetObject.HistoricalData() #Get the available historical data for Apple Inc.
```

```python
AssetObject.Informations() #Get the available information for Apple Inc.
```

```python
AssetObject.RealTime() #Get real-time data for Apple Inc.
```

```python
DataType = 'raw' #Set the data type to raw.
ExchangeCode = 'oprac' #Set the exchange code to 'oprac' for Composite.
Strategy: str = "callput" #Fetch data for calls & puts.

RawOptions = AssetObject.Options(DataType='raw') #Get the available options data for Apple Inc.
RawOptions.Data.head() #Display the first 5 rows of the data.
```

```python
RiskFreeRate = r #Set the risk-free rate to the Central Bank Policy Rate for the United States.
```

```python
ProcessedOptions = AssetObject.Options('processed',ExchangeCode,Strategy,RiskFreeRate) #Get the processed options data for Apple Inc.
ProcessedOptions.Data.head() #Display the first 5 rows of the data.
```

```python
ProcessedOptions.Plot2D(Type='call') #Plot the price of the options data for Apple Inc.
```

```python
ProcessedOptions.Plot2D('Delta', 'Call', StrikeRange=(150,300), YToExpiryRange=[0,1]) #Plot Delta for call options with a strike price between 150 and 300 and a time to expiry between 0 and 1.
```

```python
ProcessedOptions.Plot2D(Type='put') #Plot the price of the options data for Apple Inc.
```

```python
ProcessedOptions.Plot3D('Delta', 'Put', StrikeRange=(150,300), YToExpiryRange=[0,1] ,ViewAngle=(15,60)) #Plot Delta for put options with a strike price between 150 and 300 and a time to expiry between 0 and 1.
```

```python
ProcessedOptions.Plot3D('Gamma', StrikeRange=(150,300), YToExpiryRange=[0,1] ,ViewAngle=(15,30)) #Plot Gamma for options with a strike price between 150 and 300 and a time to expiry between 0 and 1
```

```python
ProcessedOptions.Plot3D('Vega', StrikeRange=(150,300), YToExpiryRange=[0,1] ,ViewAngle=(15,150)) #Plot vega for put options with a strike price between 150 and 300 and a time to expiry between 0 and 1.
```

```python
ProcessedOptions.Plot3D('iv', Type='call', ViewAngle=(15,45)) #Plot all the greeks for call options data for Apple Inc.
```
